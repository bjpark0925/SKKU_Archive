# 1. Pandas를 사용하여 데이터를 로드하고, 처음 15개의 행을 출력

```python
import pandas as pd

#Pandas를 사용하여 데이터를 로드
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
df = pd.read_csv(url, names=['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship',
                'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income'])

#처음 15개의 행을 출력
df[:15]
```

# 2. 데이터셋에서 범주형 변수만 선택하여 별도의 DataFrame으로 저장 후 확인

```python
import numpy as np

# select_dtypes 이용하여 수치형 col 제외한 범주형 col 선택
categorical_columns = df.select_dtypes(exclude=np.number).columns.tolist()
# column은 ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'native-country', 'income']

# 범주형 변수만 선택한 새로운 DataFrame 생성
cat_df = df[categorical_columns]

# print(cat_df.head())
cat_df[:15]
```

# 3. workclass 변수를 레이블 인코딩으로 변환

```python
from sklearn.preprocessing import LabelEncoder

# LabelEncoder 객체 생성
label_encoder = LabelEncoder()

# workclass 열을 레이블 인코딩
cat_df['workclass'] = label_encoder.fit_transform(cat_df['workclass'])

print(cat_df[['workclass']])
```

# 4. education 변수를 원-핫 인코딩으로 변환

```python
# pandas 사용
# education 변수를 원-핫 인코딩
# get_dummies로 각 고유한 education 범주가 새로운 열로 표시됨. prefix는 새로운 열의 이름에 'education_' 접두사 추가
cat_df = pd.get_dummies(cat_df, columns=['education'])

# 결과 확인
print(cat_df)
```

# 5. 변환된 dataframe 처음 15개 행 & 마지막 15개 행 출력

```python
print(cat_df.head(15))
print(cat_df.tail(15))
```
