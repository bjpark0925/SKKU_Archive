```python
# 필요한 라이브러리 임포트
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# iris 데이터셋 로드
iris = datasets.load_iris()
X = iris.data
y = iris.target

# KMeans 클러스터링
kmeans = KMeans(n_clusters=3, n_init='auto', random_state=42)
y_kmeans = kmeans.fit_predict(X)
y_kmeans
```

```python
# 클러스터링 결과 시각화
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=50, c='red', label='Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=50, c='blue', label='Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s=50, c='green', label='Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            marker='^', s=120, c='yellow', edgecolor='k', label='Centroids')
plt.title('KMeans Clustering of Iris Dataset')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.legend()
plt.show()

# 실루엣 점수 출력
score = silhouette_score(X, y_kmeans)
print(f"Silhouette Score: {score:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# iris 데이터 로드
iris = datasets.load_iris()
X = iris.data

# 데이터 표준화
X = StandardScaler().fit_transform(X)

# DBSCAN 적용
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(X)
print(clusters)
```

```python
# 결과 시각화
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("DBSCAN Clustering of Iris Dataset")
plt.show()
```

```python
# 결과 시각화
plt.scatter(X[:, 2], X[:, 3], c=clusters, cmap='plasma')
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
plt.title("DBSCAN Clustering of Iris Dataset")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

# iris 데이터 로드
iris = datasets.load_iris()
X = iris.data

# 계층적 클러스터링의 연결(linkage) 계산
linked = linkage(X, 'ward')

# 덴드로그램(dendrogram) 시각화
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title("Hierarchical Clustering Dendrogram of Iris Dataset")
plt.show()
```

```python
# AgglomerativeClustering을 사용하여 클러스터링 수행
agg_clustering = AgglomerativeClustering(n_clusters=3, linkage='ward')
clusters = agg_clustering.fit_predict(X)

# 결과 시각화
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='inferno')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("Agglomerative Clustering of Iris Dataset")
plt.show()
```

```python
# 결과 시각화
plt.scatter(X[:, 2], X[:, 3], c=clusters, cmap='autumn')
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
plt.title("Agglomerative Clustering of Iris Dataset")
plt.show()
```

```python
# Affinity Propagation

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import AffinityPropagation
from sklearn.metrics import silhouette_score

# iris 데이터 로드
iris = datasets.load_iris()
X = iris.data

# Affinity Propagation 적용
clustering = AffinityPropagation(random_state=42).fit(X)
labels = clustering.labels_

# 클러스터 결과 출력
print("클러스터 레이블:", np.unique(labels))

# silhouette_score 계산
silhouette_avg = silhouette_score(X, labels)
print("Silhouette Score:", silhouette_avg)
```

```python
# 시각화
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow', edgecolor='k', s=50)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("Affinity Propagation Clustering")
plt.show()
```

```python
# 시각화
plt.scatter(X[:, 2], X[:, 3], c=labels, cmap='rainbow', edgecolor='k', s=50)
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
plt.title("Affinity Propagation Clustering")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import SpectralClustering
from sklearn.metrics import silhouette_score

# iris 데이터 로드
iris = datasets.load_iris()
X = iris.data

# Spectral Clustering 적용
clustering = SpectralClustering(n_clusters=3, assign_labels="discretize", random_state=42).fit(X)
labels = clustering.labels_

# 클러스터 결과 출력
print("클러스터 레이블:", np.unique(labels))

# silhouette_score 계산
silhouette_avg = silhouette_score(X, labels)
print("Silhouette Score:", silhouette_avg)
```

```python
# 시각화
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow', edgecolor='k', s=50)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("Spectral Clustering")
plt.show()
```

```python
# 시각화
plt.scatter(X[:, 2], X[:, 3], c=labels, cmap='rainbow', edgecolor='k', s=50)
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
plt.title("Spectral Clustering")
plt.show()
```
