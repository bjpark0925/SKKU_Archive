```python
# 필요한 라이브러리 불러오기
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_iris
import pandas as pd

# Iris 데이터셋 불러오기
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]

# 독립 변수와 종속 변수 설정
X = iris_df[['sepal length (cm)', 'sepal width (cm)']]
y = iris_df[['petal length (cm)', 'petal width (cm)']]

# 다중 선형 회귀 모델 생성 및 학습
model = LinearRegression()
model.fit(X, y)

# 예측
y_pred = model.predict(X)

# 예측 결과를 DataFrame으로 변환
y_pred_df = pd.DataFrame(y_pred, columns=['Predicted petal length (cm)', 'Predicted petal width (cm)'])

# 실제 값과 예측 값 출력
print("Actual values:")
print(y.head())
print("\nPredicted values:")
print(y_pred_df.head())

# 회귀 계수 및 절편 출력
print(f"\nCoefficients: {model.coef_}")
print(f"Intercept: {model.intercept_}")
```

```python
# 필요한 라이브러리 불러오기
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_iris
import pandas as pd

# Iris 데이터셋 불러오기
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# 독립 변수와 종속 변수 설정
X = iris_df[['sepal length (cm)', 'sepal width (cm)']]
y = iris_df[['petal length (cm)', 'petal width (cm)']]

# 다중 선형 회귀 모델 생성 및 학습
model = LinearRegression()
model.fit(X, y)

# 예측
y_pred = model.predict(X)

# 예측 결과를 DataFrame으로 변환
y_pred_df = pd.DataFrame(y_pred, columns=['Predicted petal length (cm)', 'Predicted petal width (cm)'])

# 시각화
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Sepal length에 따른 Petal length & width
axes[0].scatter(X['sepal length (cm)'], y['petal length (cm)'], color='b', label='Actual petal length')
axes[0].scatter(X['sepal length (cm)'], y_pred_df['Predicted petal length (cm)'], color='r', label='Predicted petal length')
axes[0].set_title('Sepal Length vs Petal Length')
axes[0].set_xlabel('Sepal Length (cm)')
axes[0].set_ylabel('Petal Length (cm)')
axes[0].legend()

# Sepal width에 따른 Petal length & width
axes[1].scatter(X['sepal width (cm)'], y['petal width (cm)'], color='g', label='Actual petal width')
axes[1].scatter(X['sepal width (cm)'], y_pred_df['Predicted petal width (cm)'], color='y', label='Predicted petal width')
axes[1].set_title('Sepal Width vs Petal Width')
axes[1].set_xlabel('Sepal Width (cm)')
axes[1].set_ylabel('Petal Width (cm)')
axes[1].legend()

plt.show()
```
