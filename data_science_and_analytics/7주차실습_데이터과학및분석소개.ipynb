````python
#데이터 불러오기
#index_col 매개변수는 DataFrame의 인덱스로 사용할 열을 지정
data=pd.read_excel('/content/fine_dust.xlsx',index_col='area')
````

````python
#data의 처음 5개 행 확인하기
data.head()
````

````python
#data의 정보 확인하기
data.info()
````

````python
#2018년도 데이터만 추출하여 data2018에 저장하기
data2018=data[2018]
data2018
````

````python
#그래프의 사이즈는 15,4로 생성하기
plt.figure(figsize=(15,4))
#data2018의 데이터를 선 그래프로 나타내되 마커는 'o', 선 색은 파랑으로 설정하기
plt.plot(data2018,color='b',marker='o')
#그래프에 그리드 설정하기
plt.grid()
````

참고



```
plt.legend()는 Matplotlib 라이브러리의 함수로, 그래프 내에 범례(legend)를 추가하는 데 사용된다. 범례는 그래프에 여러 데이터 세트나 카테고리가 있을 때 각 데이터 세트나 카테고리를 구분하기 위한 설명 또는 식별자를 제공한다.

기본 사용법:
-데이터를 플로팅할 때 label 인자를 사용하여 각 데이터 세트의 이름을 지정한다.
-그 후, plt.legend() 함수를 호출하여 그래프에 범례를 추가한다
```



````python
#그래프의 사이즈는 15,4로 생성하기
plt.figure(figsize=(15,4))
#2016년도부터 2019년도까지 데이터를 불러와서 선 그래프로 나타내기
#이때, 마커는 's'로 표시하고, 연도를 범례로 표시하기

for year in range(2016,2020):
  chartdata=data[year]
  plt.plot(chartdata,marker='s',label=year)
plt.legend()
plt.grid()
````

````python
#2017년도 데이터를 추출하여 data2017에 저장
data2017=data[2017]
print(data2017.info())
#그래프의 사이즈는 15,4
plt.figure(figsize=(15,4))
#막대그래프로 표시하기(x축:지역, y축:data2017값, 색성은 그린)
plt.bar(data2017.index, data2017,color='g')
plt.ylim(35,55) #y축은 35~55까지 표시
plt.grid()
````

````python
#np.arange(4)는 [0, 1, 2, 3]을 생성
#이는 초기 막대 그래프 위치를 지정하기 위한 인덱스 값
index=np.arange(4)
#'Gyeonggi'~'Daegu'까지 2016년도부터 2019데이터 추출하여 data4에 저장하기
data4=data.loc['Gyeonggi':'Daegu',2016:2019]

for year in range(2016,2020):
  chartdata=data4[year]
  #index는 막대의 위치, chartdata는 막대의 높이(값), width는 막대의 너비, label은 범례에 표시될 라벨
  plt.bar(index,chartdata,width=0.2,label=year)
  index=index+0.2
plt.ylim(35,55)
# xticks: x축의 눈금 위치와 레이블을 설정
plt.xticks(index-0.5,['Gyeonggi','Incheon','Busan','Daegu'])
plt.legend()
````
