````python
#다변량 데이터 시각화하기
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
o = OrdinalEncoder()
l = LabelEncoder()
for i in df.columns:
  if df[i].dtypes=='O' and i!='Attrition':
    df[i]=o.fit_transform(df[i].values.reshape(-1,1))

df['Attrition']=l.fit_transform(df['Attrition'])
````

<결론>

일한 회사의 수가 적을수록 이직 의사가 높다.  
해당 회사가 첫 직장인 직원은 입사 2년차부터 이직을 고려한다.  
일한 회사의 수가 3~8개인 직원은 입사와 동시에 이직을 고려한다.  
일한 회사의 수가 8개 초과인 직원은 이직 의사가 상대적으로 낮다.  
일한 회사의 수가 4-8개인 직원은 5년차와 10년차에 한번씩 이직을 고려하지 않는 시점이 있다.  

<해결방법>

일한 회사 개수별 근속 연수에 따라 직원이 이직하지 않도록 인센티브를 제공한다. 예를 들어 일한 회사의 수가 2인 직원은 0년차와 1년차에 이직을 고려하지 않으므로 이 직원에겐 입사 2년차부터 인센티브 제도를 도입한다.
