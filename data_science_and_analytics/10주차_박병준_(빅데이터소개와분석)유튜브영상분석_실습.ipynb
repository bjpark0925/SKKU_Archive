```python
#필요한 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#데이터 읽어와서 확인하기
df=pd.read_csv('/content/KR_youtube_trending.csv')
df.head()
```

```python
#데이터 정보 확인
df.info()
```

```python
#'publishedAt'속성을 datetime으로 변환 0000-00-00 00:00:00
df.publishedAt=pd.to_datetime(df['publishedAt'], format='%Y-%m-%dT%H:%M:%SZ' )
df.info()
```

```python
def getDatetime(x):
  if 4<x.hour<=10:  #5시~10시까지는 'morning'
    return 'morning'
  elif 10<x.hour<=17: #11시~17시까지는 'afternoon'
    return 'afternoon'
  elif 17<x.hour<=22: #18시~22까지는 'evening'
    return 'evening'
  else:   #나머지 시간대는 'night'
    return 'night'

#'datetime'속성에 'pupublishedAt'에 대한 시간대 저장하기
df['datetime']=df['publishedAt'].apply(getDatetime)
```

```python
df['datetime'].value_counts()
```

```python
#trending_date을 '%Y-%m-%dT%H:%M:%SZ'형식인 datetime으로변화
df['trending_date']=pd.to_datetime(df['trending_date'], format='%Y-%m-%dT%H:%M:%SZ')
df.info()
```

```python
#인기영상리스트로 게재(trending_date)되기까지 걸리는 시간
df['time']=df.trending_date-df.publishedAt
```

```python
#선그래프로 나타내기
df['time'].value_counts().plot()
```

```python
#조회수('view_count')가 가장 많은 유튜브영상 정보 확인:loc,idxmax()
df.loc[df['view_count'].idxmax()]
```

카테고리별로 동영상을 나누지 않고, 전체 동영상을 대상으로 인기동영상으로 개제되는 데까지 걸리는 시간의 평균을 계산해보았다.

```python
# time 값을 days 단위로 분할
df['time_days'] = df['time'].dt.days

#인기 동영상이 되기까지 걸리는 시간 분포 시각화
plt.hist(df['time_days'], bins=20, color='skyblue')
plt.xlabel('Time (days)')
plt.ylabel('Frequency')
plt.title('Distribution of time_days')

plt.show()
```

분포를 토대로, 인기 동영상으로 개제되는 데까지 걸리는 시간이 짧으면 hottest로 그룹화하고, 길면 cool로 그룹화했다. 개제되는 시간이 짧은 순서대로 hottest, hot, warm, cool로 명명했다.

```python
# 인기 동영상 개제까지 걸리는 시간에 따라 그룹을 나눔
def splitTime(x):
  if 0 < x <= 2:  # 인기 동영상 개제까지 걸리는 시간이 2일 이하인 경우 'hottest'
    return 'hottest'
  elif 2 < x <= 4:  # 2일 초과 4일 이하는 'hot'
    return 'hot'
  elif 4 < x <= 6:  # 4일 초과 6일 이하는 'warm'
    return 'warm'
  else:  # 나머지는 'cool'
    return 'cool'

# 'time_days' 열을 기반으로 'splitTime' 열 생성
df['splitTime'] = df['time_days'].apply(splitTime)
```

그룹에 따른 조회수, 좋아요, 싫어요, 댓글수를 분석해보았다. 조회수와 좋아요 수, 싫어요 수 모두 인기 동영상 개제까지 걸리는 시간과 반비례하는 경향을 보였다. 인기 동영상 개제까지 걸리는 시간이 짧을수록 조회수나 좋아요 수, 싫어요 수가 적다는 의미이다. 댓글수도 비슷한 경향을 보였으나, hottest가 hot보다 더 높은 댓글 수를 기록했다. 가장 짧은 시간 안에 인기 동영상으로 개제된 영상일수록 댓글에서 자신의 의사를 더 많이 표현하거나 열띤 토론이 이루어진다는 것을 볼 수 있다.

```python
# hottest, hot, warm, cool 영상별 조회수, 좋아요, 싫어요, 댓글 평균 수
df_hottest=df.groupby('splitTime').agg({'view_count':'mean', 'likes':'mean', 'dislikes':'mean', 'comment_count':'mean'})
sns.barplot(data=df_hottest, x=df_hottest.index, y='view_count')
```
