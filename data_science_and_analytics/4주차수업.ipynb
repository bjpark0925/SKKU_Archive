```python
import pandas as pd
decade=[1960,1960,1960,1960,1960,1960,1970,1970,1970,1970,1970,1970,1980,1980,1980,1980]
gender=['F','M','F','M','F','M','F','M','F','M','F','M','F','M','F','M']
births=[4192,4198,4710,4850,4646,5092,4800,4934,4126,4324,4758,5076,5070,5296,4798,5096]
dataset=pd.DataFrame({'decade':decade,'gender':gender,'births':births})
dataset
```

```python
#출생년도별 성별 인원수 출력하기
dataset.groupby(by='decade',as_index=False).sum()
```

```python
data=pd.read_csv('/content/movies_train.csv')
```

```python
data
```

```python
#장르별 관객수
data.groupby(by=['genre'],as_index=False).sum()
```

```python
#장르별 최대 관객수,영화제목과 감독이름
#무엇이 문제일까?
temp=data.groupby(['genre'],as_index=False)['box_off_num','title','director'].max()
temp.sort_values(by='box_off_num',ascending=False)
```

```python
#장르별 영화편수 : size()
temp=data.groupby(['genre'],as_index=False).size()
temp.sort_values(by='size',ascending=False)
```
